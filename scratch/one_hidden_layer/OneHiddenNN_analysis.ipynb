```python
import numpy as np
import matplotlib.pyplot as plt

from lib.utils import load_planar_dataset
from lib.OneHiddenNN import OneHiddenLayerClassifier

np.random.seed(1)
```

```python
X, y = load_planar_dataset()
```

```python
X.shape
```

```python
y.shape
```

```python
# Visualize the data:
plt.scatter(X[:, 0], X[:, 1], c=y, s=40, cmap=plt.cm.Spectral);
```

```python
nn = OneHiddenLayerClassifier(n_hidden1=4, max_iter=10000, learning_rate=1.2)
```

```python
nn.fit(X, y)
```

```python
nn.plot_decision_boundary(X, y)
```

```python
for i, cost in enumerate(nn.costs):
    if i % 1000 == 0:
        print(cost)
```

```python
accuracy = nn.get_accuracy(X, y)
accuracy
```

```python
plt.figure(figsize=(16, 32))
hidden_layer_sizes = [1, 2, 3, 4, 5, 20, 50]
for i, n_hidden1 in enumerate(hidden_layer_sizes):
    plt.subplot(5, 2, i+1)
    plt.title('Hidden Layer of size {}'.format(n_hidden1))
    nn = OneHiddenLayerClassifier(n_hidden1=n_hidden1, max_iter=10000, learning_rate=1.2)
    nn.fit(X, y)
    nn.plot_decision_boundary(X, y)
    accuracy = nn.get_accuracy(X, y)
    print ("Accuracy for {} hidden units: {}".format(n_hidden1, accuracy))
```
